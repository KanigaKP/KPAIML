```python
board = [' '] * 9

def print_board():

    print(f"[ {board[0]} , {board[1]} , {board[2]} ]")
    print(f"[ {board[3]} , {board[4]} , {board[5]} ]")
    print(f"[ {board[6]} , {board[7]} , {board[8]} ]")

def ai_move():

    for i in range(9):

        if board[i] == ' ':

            board[i] = 'O'

            break

for i in range(4):

    pos = int(input("Enter position (0-8): "))

    if board[pos] == ' ':

        board[pos] = 'X'

        ai_move()

        print_board()

    else:

        print("Position already filled")

print("Final Board :")

print_board()
```

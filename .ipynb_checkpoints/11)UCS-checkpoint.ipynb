```python
import heapq

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}

def ucs(start, goal):

    queue = [(0, start, [])]

    while queue:

        cost, node, path = heapq.heappop(queue)

        path = path + [node]

        if node == goal:

            return path, cost

        for n, w in graph[node]:

            heapq.heappush(queue, (cost + w, n, path))

path, cost = ucs('A', 'F')

print("Path :", path)

print("Cost :", cost)

print("Arrow :", " -> ".join(path))
```

```python
import heapq

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2)],
    'C': [('F', 1)],
    'D': [],
    'F': []
}

heuristic = {
    'A': 5,
    'B': 2,
    'C': 1,
    'D': 0,
    'F': 0
}

def astar(start, goal):

    queue = [(0, start, [])]

    cost = {
        start: 0
    }

    while queue:

        f, node, path = heapq.heappop(queue)

        path = path + [node]

        if node == goal:

            return path, cost[node]

        for n, w in graph[node]:

            g = cost[node] + w

            f = g + heuristic[n]

            heapq.heappush(queue, (f, n, path))

            cost[n] = g

path, cost = astar('A', 'F')

print("Path :", path)

print("Cost :", cost)

print("Arrow :", " -> ".join(path))
```

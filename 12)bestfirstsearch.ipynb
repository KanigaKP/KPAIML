```python
import heapq

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

heuristic = {
    'A': 5,
    'B': 3,
    'C': 4,
    'D': 1,
    'E': 2,
    'F': 0
}

def best_first(start, goal):

    queue = [(heuristic[start], start, [])]

    while queue:

        h, node, path = heapq.heappop(queue)

        path = path + [node]

        if node == goal:

            return path

        for n in graph[node]:

            heapq.heappush(queue, (heuristic[n], n, path))

path = best_first('A', 'F')

print("Path :", path)

print("Arrow :", " -> ".join(path))
```

```python
values = {
    'A': 1,
    'B': 3,
    'C': 5,
    'D': 7,
    'E': 6
}

graph = {
    'A': ['B'],
    'B': ['C'],
    'C': ['D'],
    'D': ['E'],
    'E': []
}

def hill_climb(start):

    current = start

    path = [current]

    while graph[current]:

        next_node = graph[current][0]

        if values[next_node] > values[current]:

            current = next_node

            path.append(current)

        else:

            break

    return path

path = hill_climb('A')

print("Path :", path)

print("Arrow :", " -> ".join(path))
```

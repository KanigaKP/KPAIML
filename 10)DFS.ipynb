```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def dfs(node, visited):

    if node not in visited:

        visited.append(node)

        for n in graph[node]:

            dfs(n, visited)

    return visited

result = dfs('A', [])

print("DFS :", result)

print("Path :", " -> ".join(result))
```
